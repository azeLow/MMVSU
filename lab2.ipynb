```python
import numpy as np
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import PolynomialFeatures


# Создаем первый график
year = []
for i in range (1991,2021):
  year.append(i)
population = [1061000, 1059000, 1055000, 1049000, 1044000, 1039000, 1035000, 1031000, 1027000, 1023000, 1001653, 994600, 989500, 993300, 990200, 987234, 985794, 991162, 991882, 1000672, 1013887, 1026477,	1036469, 1041876, 1048005, 1051583, 1053938, 1055397, 1065397, 1075397]

# Создаем массивы numpy для годов и населения
year = np.array(year).reshape(-1, 1)
population = np.array(population)

poly_features = PolynomialFeatures(degree=2, include_bias=False)
year_poly = poly_features.fit_transform(year)

# Создаем модель квадратичной регрессии и обучаем ее на данных
model = LinearRegression()
model.fit(year_poly, population)

# Получаем предсказанные значения населения для всех годов
predicted_population = model.predict(year_poly)

# Отображаем график с данными и предсказанными значениями
pylab.plot(year, population, label='Данные')
pylab.plot(year, predicted_population, label='Квадратичная регрессия')
pylab.title('Изменение населения')
pylab.xlabel('Год')
pylab.ylabel('Население')
pylab.legend()
pylab.show()
```
