```python
import pandas as pd 
import matplotlib.pyplot as plt
import numpy as np
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import PolynomialFeatures

url = 'https://raw.githubusercontent.com/azeLow/MMVSU/main/perm.csv'
df = pd.read_csv(url)

years = np.array(df['Year']).reshape(-1, 1)
population = np.array(df['Population'])

poly_features = PolynomialFeatures(degree = 2, include_bias = False)
year_poly = poly_features.fit_transform(years)

# Создаем модель регрессии и обучаем ее на данных
model = LinearRegression()
model.fit(year_poly, population)

# Получаем предсказанные значения населения для всех годов
predicted_population = model.predict(year_poly)

# Отображаем график с данными и предсказанными значениями
plt.plot(years, population)
plt.plot(years, predicted_population)
plt.plot(years, population)


```
